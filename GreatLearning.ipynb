# **all the teams which have started playing between 1930-1980**

```python
data['Unnamed: 17']= data['Unnamed: 17'].astype(str)
Debut_year= data[data['Unnamed: 17'].str[:4].between('1930','1980')]
Debut_year[['Unnamed: 1', 'Unnamed: 17']]
```

# **list of teams which came Top 5 in terms of points**

```python
top_five= data[1:][['Unnamed: 1', 'Unnamed: 3']]
top_five['Unnamed: 3']= top_five['Unnamed: 3'].astype(int)
top_five.sort_values(by='Unnamed: 3', ascending=False, inplace=True)
top_five.head()
```

# **Goal_diff_count**

```python
goal_diff= data[1:][['Unnamed: 1','Unnamed: 8', 'Unnamed: 9']]
goal_diff['Unnamed: 8']= goal_diff['Unnamed: 8'].astype(int)
goal_diff['Unnamed: 9']= goal_diff['Unnamed: 9'].astype(int)



def Goal_diff_count():
  goal_diff['Goal_difference']= goal_diff['Unnamed: 8']-goal_diff['Unnamed: 9']
  return goal_diff[['Unnamed: 1','Goal_difference']]

goal= Goal_diff_count()



```

# **maximum and minimum goal difference**

```python
goal.sort_values(by='Goal_difference', ascending=False, inplace=True)
goal
```

```python
goal.head(1)
```

# **Winning Percent**

```python
data["Unnamed: 4"]= data[1:]["Unnamed: 4"].astype(int)
data["Unnamed: 5"]= data[1:]["Unnamed: 5"].astype(int)

data["Win_Percentage"]= (data["Unnamed: 5"]/data["Unnamed: 4"])*100
data["Win_Percentage"].fillna(0,inplace=True)

w= data[1:][["Unnamed: 1", "Win_Percentage"]]
w
```

```python
w.sort_values(by="Win_Percentage", ascending=False, inplace=True)
w.head(5)
```
